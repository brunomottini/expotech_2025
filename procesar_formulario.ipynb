```python
import csv
from docx import Document
import re
```

```python
# Ruta del archivo CSV
file_path = "./retriever_documents/formulario.csv"
```

```python
# Crear un objeto Document de python-docx
doc = Document()

# Agregar un título inicial al documento
doc.add_heading('Resumen de Formularios', level=1)

with open(file_path, 'r', ) as f:
    reader = csv.DictReader(f)
    for idx, row in enumerate(reader, start=1):
        company_name = row.get("Nombre de la Compañía", "").strip()
        productos_servicios = row.get("¿Cuáles son los principales productos o servicios que ofrece su empresa?", "").strip()
        objetivos = row.get("¿Qué espera lograr al participar de ExpoTech 2025?", "").strip()
        networking = row.get("¿Desea ser contactado para Networking?", "").strip()
        tipo_empresas_busca = row.get("¿Qué tipo de empresas o socios busca conocer?", "").strip()
        faqs = row.get("¿Qué preguntas frecuentes recibe su empresa de clientes potenciales? (FAQ)", "").strip()
        contactos = row.get("Contactos durante la ExpoTech", "").strip()

        # Crear un subtítulo para cada compañía
                # Crear título de sección
        title_text = f"{idx}. {company_name if company_name else 'Empresa Sin Nombre'}"
        doc.add_heading(title_text, level=2)

        # Procesar productos/servicios (asumiendo un formato similar al ejemplo)
        # Intentamos extraer registros del tipo: "Tipo: X, Nombre: Y, Descripción: Z, Público Objectivo: W, ¿Hay algún caso de éxito?: Q"
        items = productos_servicios.split('\n')
        pattern = re.compile(r"Tipo:\s*(?P<tipo>[^,]+),\s*Nombre:\s*(?P<nombre>[^,]+),\s*Descripción:\s*(?P<descripcion>[^,]+),\s*Público Objectivo:\s*(?P<publico>[^,]+),\s*¿Hay algún caso de éxito de este producto o servicio que desee destacar\?:\s*(?P<exito>.*)")

        prod_serv_summary = ""
        for item in items:
            match = pattern.search(item)
            if match:
                tipo = match.group('tipo').strip()
                nombre = match.group('nombre').strip()
                descripcion = match.group('descripcion').strip()
                publico = match.group('publico').strip()
                exito = match.group('exito').strip()

                prod_serv_summary += f"- {tipo.capitalize()}: {nombre}\n  Descripción: {descripcion}\n  Público Objetivo: {publico}\n  Caso de Éxito: {exito}\n\n"
            else:
                # Si no matchea el patrón, quizás es una entrada vacía o formato distinto
                if item.strip():
                    prod_serv_summary += f"- {item.strip()}\n"
        
        # Procesar FAQs (ejemplo: "Pregunta: X, Respuesta: Y\nPregunta: A, Respuesta: B")
        faq_pattern = re.compile(r"Pregunta:\s*(?P<pregunta>[^,]+),\s*Respuesta:\s*(?P<respuesta>.*)")
        faqs_entries = faqs.split('\n')
        faqs_summary = ""
        for faq_item in faqs_entries:
            faq_match = faq_pattern.search(faq_item)
            if faq_match:
                pregunta = faq_match.group('pregunta').strip()
                respuesta = faq_match.group('respuesta').strip()
                faqs_summary += f"Pregunta: {pregunta}\nRespuesta: {respuesta}\n\n"
            else:
                if faq_item.strip():
                    faqs_summary += f"{faq_item.strip()}\n"

         # Antes de agregar todo al documento, imprimimos el texto en consola:
        print(title_text)
        if prod_serv_summary:
            print("Productos / Servicios:")
            print(prod_serv_summary)
        if faqs_summary:
            print("Preguntas Frecuentes (FAQ):")
            print(faqs_summary)
        print("Información General:")
        print(resumen)
        print("="*50)  # Separador para cada empresa

        # Crear el texto resumen para esta empresa
        # Incluimos objetivos, networking, tipo de empresas buscadas, contactos, etc.
        resumen = f"**Objetivos en ExpoTech 2025:** {objetivos or 'No especificado'}\n"
        resumen += f"**Desea ser contactado para Networking:** {networking or 'No especificado'}\n"
        resumen += f"**Tipo de empresas o socios que busca conocer:** {tipo_empresas_busca or 'No especificado'}\n"
        resumen += f"**Contactos durante la ExpoTech:** {contactos or 'No especificado'}\n\n"
        
        if prod_serv_summary:
            doc.add_heading("Productos / Servicios:", level=3)
            doc.add_paragraph(prod_serv_summary)

        if faqs_summary:
            doc.add_heading("Preguntas Frecuentes (FAQ):", level=3)
            doc.add_paragraph(faqs_summary)

        doc.add_heading("Información General:", level=3)
        doc.add_paragraph(resumen)
```
